# Analog Signal Defination

```python
def signalsGen(Fs,Fc,StopTime,Amp):
    global t
    sampling_period = 1/Fs                   # seconds per sample
    sample_number  = StopTime / sampling_period; # Sample Number
    t = np.linspace(0, StopTime, sample_number) 
    Chirp = chirp(t, f0=2.5, f1=12.5, t1=StopTime, method='linear')*Amp+Amp
    ex = np.exp(-t*0.5)*Amp+Amp
    S = (np.sin(Fc*t))*Amp+Amp
    S_sq = (np.sin(Fc*t**2))*Amp+Amp
    C = (np.cos(Fc*t))*Amp+Amp
    Sq = signal.square(Fc * t)*Amp+Amp
    Saw = signal.sawtooth(Fc * t)*Amp+Amp
    return (Chirp, ex,S, S_sq,C,Sq,Saw)
```

```python
Sig = signalsGen(Fs = 1000, StopTime= 10, Amp = 5, Fc=15)  
```
